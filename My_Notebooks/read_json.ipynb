```python
iterables = []
hash_table = {}
for entry in new_info.portfolio_companies:
    
    try:
        if len(entry) > 1:
            for i in entry:
                for key in i.keys():
                    if key not in hash_table:
                        hash_table[key] = [i[key]]
                    elif key in hash_table:
                        hash_table[key].append(i[key])
                    
        else:
            for key1 in entry[0].keys():
                if key1 not in hash_table:
                    hash_table[key1] = [entry[0][key1]]
                elif key in hash_table:
                    hash_table[key1].append(entry[0][key1])
    except TypeError:
        continue

    
```
