## Eigendecomposition
An eigenvector of a squre matrix A is a nozero vector v such that multiplcation by A alters only the scale of v.

The scalar lambda is known as the eigenvalue corresponding to this eigenvector

find the eigenvalue and eigenvecotor for a simple square matrix

```python
A = np.diag(np.arange(1,4))
A
```

```python
eigenvalues, eigenvectors = np.linalg.eig(A)
```

```python
eigenvalues
```

```python
eigenvectors
```

the eigenvalue w[i] corresponds to eigenvector v[:1]

```python
print('Eigenvalue: {}'.format(eigenvalues[1]))
```

```python
print('Eigenvector: ')
print('{}'.format(eigenvectors[:,1]))
```

We can easily double check that these are the correct eigenvalues and eigenvectors by plugging them back into the eigendecomposition equation. The eigen composition of A is given by where lambda is equal to a vector of the eigen values, and V is a matrix with one eigenvector per column

## verify eigendecomposition

```python
np.diag(eigenvalues)
```

```python
np.linalg.inv(eigenvectors)
```

```python
matrix = np.matmul(np.diag(eigenvalues),np.linalg.inv(eigenvectors))
```

```python
matrix
```

```python
output = np.matmul(eigenvectors, matrix).astype(np.int)
```

```python
output
```

import necessary maplotlib libraries

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

plot the eigenvectors

```python
origin = [0,0,0]

fig = plt.figure(figsize=(18,10))
fig.suptitle('Effects of eigenvalues and eigenvectors')
axl = fig.add_subplot(121,projection = '3d')
axl.quiver(origin,origin,origin, eigenvectors[0,:],eigenvectors[1,:],eigenvectors[2,:],color ='k')
axl.set_xlim([-3,3])
axl.set_ylim([-3,3])
axl.set_zlim([-3,3])
axl.set_ylabel('X axis')
axl.set_xlabel('Y axis')
axl.set_zlabel('Z axis')
axl.view_init(15,30)
axl.set_title('Before Multiplication')

new_eig = np.matmul(A,eigenvectors)
ax2 = plt.subplot(122,projection = '3d')
ax2.quiver(origin,origin,origin, new_eig[0,:],new_eig[1,:],new_eig[2,:],color ='k')
ax2.plot((eigenvalues[0]*eigenvectors[0]),(eigenvalues[1]*eigenvectors[1]),(eigenvalues[2]*eigenvectors[2]),'rX')
ax2.set_xlim([-3,3])
ax2.set_ylim([-3,3])
ax2.set_zlim([-3,3])
ax2.set_ylabel('X axis')
ax2.set_xlabel('Y axis')
ax2.set_zlabel('Z axis')
ax2.view_init(15,30)
ax2.set_title('After Multiplication')
plt.show()
```
