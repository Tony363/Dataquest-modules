+ Calculate each team's average ERA from 2011 to 2015 and list the result as a table,where each column is a year, each row is a team

```python
modern_years = df.loc[(df.yearID >= 2011) & (df.yearID <=2015),:]
```

```python
df_list = pd.concat([df.loc[df.yearID == year,['teamID','ERA']].groupby(['teamID']).mean() for year in modern_years.yearID.unique()],axis=1)
```

```python
df_list.columns = [year for year in modern_years.yearID.unique()]
```

```python
df_list
```

+ Calculate the average ERAs for American League and National leauge from 2011 to 2015 and list te result as a table, wjere each column is a league and each row is a year

```python
df_American = pd.concat([df.loc[(df.yearID == year) & (df.lgID == 'AL'),['teamID','ERA','lgID']].groupby(['teamID']).mean() for year in modern_years.yearID.unique()],axis=1)
df_American.columns = [year for year in modern_years.yearID.unique()]
```

```python
df_American
```

```python
df_National = pd.concat([df.loc[(df.yearID == year) & (df.lgID == 'NL'),['teamID','ERA','lgID']].groupby(['teamID']).mean() for year in modern_years.yearID.unique()],axis=1)
df_National.columns = [year for year in modern_years.yearID.unique()]
```

```python
df_National
```

+ List the pitchers who had the most wins in each of the years 2011 to 2015

```python
pitcher_W = pd.concat([df.loc[df.yearID == year,['nameLast','W']].sort_values(by='W',ascending=False).reset_index().drop(columns=['W','index']) for year in modern_years.yearID.unique()],axis=1)
pitcher_W.columns = [year for year in modern_years.yearID.unique()]
```
