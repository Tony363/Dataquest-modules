```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
edges = pd.read_csv("Nodes_Data.csv")
```

```python
edges.head()
```

```python
G = nx.convert_matrix.from_pandas_edgelist(edges,source="Node1",target="Node2",edge_attr="Weight")
```

```python
G.edges()
```

```python
#pos = nx.nx_agraph.graphviz_layout(G)
pos = nx.spring_layout(G)
nx.draw(G,pos, with_labels=True)
labels = nx.get_edge_attributes(G,'Weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
```

```python
visited = []
colors = ['red'] * G.number_of_nodes()

def dfs(node):
    if not node in visited:
        print("visited",visited)
        colors[node] = 'blue'
        nx.draw(G, pos, node_color=colors, with_labels=True)
        plt.show()
        visited.append(node)
        neighbors = nx.neighbors(G,node)
        #print(list(neighbors))
        for neighbor in neighbors:
            dfs(neighbor)
```

```python
dfs(18)
```

```python
def bfs(node):
    visited, queue = set(), [node]
    while queue:
        vertex = queue.pop(0)
        if vertex not in visited:
            visited.add(vertex)
            neighbors = nx.neighbors(G,node)
            for i in neighbors:
                queue.append(i)
        print(queue)
    return visited
```

```python
kruskal_g = nx.convert_matrix.from_pandas_edgelist(edges.sort_values("Weight").head(G.number_of_nodes() - 1),
                                                   source="Node1",target="Node2",edge_attr="Weight")
```

```python
sorted_edges = sorted(G.edges(data=True), key=lambda x: x[2]['Weight'])
```

```python
sorted_edges
```

```python
Kruskal_G = nx.Graph()
for node1, node2, weight in sorted_edges:
    Kruskal_G.add_edge(node1, node2, weight=weight["Weight"])
    cycles = nx.algorithms.cycles.cycle_basis(Kruskal_G)
    #if not nx.algorithms.tree.recognition.is_tree(Kruskal_G):
    print(cycles)
    if len(cycles) > 0:
        Kruskal_G.remove_edge(node1, node2)
    if Kruskal_G.number_of_edges() == G.number_of_nodes()-1:
        break
```

```python
Kruskal_G.number_of_edges()
```

```python
for node in G.nodes():
    print(node,nx.shortest_path_length(Kruskal_G,0,node,weight = "weight"))

print("this is shortest path length")
```
