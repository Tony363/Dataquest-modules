```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn import utils
```

```python
lab_enc = preprocessing.LabelEncoder()
training_scores_encoded = lab_enc.fit_transform(y_train)
```

```python
print(utils.multiclass.type_of_target(y_train))
print(utils.multiclass.type_of_target(y_train.astype('str')))
print(utils.multiclass.type_of_target(training_scores_encoded))
```

```python
GB_classifier = GradientBoostingClassifier()
GB_classifier.fit(X_train, training_scores_encoded)#training_scores_encoded
```

```python
y_predictions = GB_classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy of Model::", accuracy_score(y_test,y_predictions))
```
