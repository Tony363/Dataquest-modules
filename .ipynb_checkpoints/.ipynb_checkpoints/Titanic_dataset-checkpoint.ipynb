```python
import pandas as pd

path = r'C:\Users\Tony\Downloads\FPedjhY58vCPJzojVYiu9c-titanic dataset.csv'

df = pd.read_csv(path, encoding='utf-8')
df 
```

```python
import pandas_profiling as pp
pp.ProfileReport(df)
```

```python
df['Sex'] = [1 if i == 'male' else 0 if i == 'female' else i for i in df['Sex']]
df['Sex']
```

```python
df['Survived'] = [0 if i == "No" else 1 if i == 'Yes' else i for i in df['Survived']]
df.head()
```

```python
df['Fare'] = df['Fare'].round(2)
df['Fare']
```

```python
df = df.sort_values(['Fare','Age'], ascending=False)
df
```

## Using sklearn GradientBoostingClassifier

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier as GBC
from sklearn import preprocessing,utils
from sklearn.linear_model import LogisticRegression
```

```python
X = df[['Pclass','Sex','Age','Siblings/Spouses_Aboard','Parents/Children_Aboard','Fare']]
X
```

```python
y = df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 0)
```

```python
X_train
```

```python
y_train
```

```python
y_test
```

```python
X_test
```

```python
gbc = GBC()
fitted = gbc.fit(X_train, y_train)
```

```python
 prediction_sk = fitted.predict(X_test)
prediction_sk
```

```python
from sklearn.metrics import accuracy_score

print('Accuracy of Model::',accuracy_score(y_test,prediction_sk))
```

## Using Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dTree_clf = DecisionTreeClassifier()
```

```python
dTree_clf.fit(X_train,y_train)
```

```python
Tree_predictions = dTree_clf.predict(X_test)
```

```python
Tree_predictions
```

```python
print('Accuracy of Model::', accuracy_score(y_test,Tree_predictions))
```

## Using XGBoost

```python
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
classifier = xgb.sklearn.XGBClassifier()
```

```python
classifier.fit(X_train,y_train)
```

```python
predictions_xgb = classifier.predict(X_test)
```

```python
predictions_xgb
```

```python
print('Accuracy of Model::', accuracy_score(y_test,predictions_xgb))
```

## Alternative Using DMatrix

```python
data_dmatrix = xgb.DMatrix(data=X,label=y)
```

```python
params = {'max_depth':2,'eta':1,'silent':1,'objective':'binary:logistic'}
```

```python
xgb_clf = xgb.train(params=params,dtrain=data_dmatrix,num_boost_round=10)
```
