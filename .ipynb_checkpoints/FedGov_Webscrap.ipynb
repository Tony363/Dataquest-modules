```python
def total_consumer_credit():
    url = urllib.request.urlopen('https://www.federalreserve.gov/releases/g19/current/default.htm').read()
    soup = BeautifulSoup(url,features="html.parser")
    Consumer_Oustanding = soup.find('table',title='Consumer Credit Outstanding')
    table = Consumer_Oustanding.find('tbody')
    row = table.find_all('tr')
    
    matrix = []
    for tr in row:
        text = str(tr.getText())
        value = text.split('\n')
        value = [i for i in value if i != '']
        matrix.append(value)
#     print(matrix) 
    
    df = pd.DataFrame(matrix)
    df = df.set_index(0)
    df.columns = [2014,2015,2016,2017,2018,'Q3r','Q4r','Q1r','Q2r','Q3r','Augr','Sepr','Octp']
    df.index = [ 'Total percent change(annual rate)',
            'Revolving_1','Nonrevolving_1',
            'Total flow(annual rate)',
            'Revolving_1',
           'Nonrevolving_2',
           'Total outstanding',
           'Revolving_2',
           'Nonrevolving_3',
           ' ',
           'Terms of Credit',
           'Commerical bank interest rates 5',
           'New car loans',
           '48-month',
           '60-month',
           'Credit card plans',
           'All accounts',
           'Accounts assessed interest',
           'Personal loans',
           '24-months',
           ' ',
           'Finance companies(new car loans)',
           'Interest rates',
           'Maturity(months)',
           'Amount financed(dollars)',
           ]
    df = df.loc[['Nonrevolving_1','Nonrevolving_2','Nonrevolving_3','Total outstanding']]
    return df
```

```python
total_consumer_credit()
```
