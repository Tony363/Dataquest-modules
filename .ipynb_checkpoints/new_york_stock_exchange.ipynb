```python
ADVC = df[df['stock_name'] == 'ADVC']
ADVC
```

```python
ADVC['date'] = pd.to_datetime(ADVC['date'],dayfirst=True)
```

```python
ADVC = ADVC.sort_values('date')
ADVC
```

```python
# Moving Averages (SMA & EWMA)


# Simple Moving Average 
def SMA(data, days): 
    sma = pd.Series(data['Close'].rolling(window=days).mean())
    name = 'SMA_' + str(days) + '_days'
    data[name] = sma
    return data

# Exponentially-weighted Moving Average 
def EWMA(data, days):
    ema = pd.Series(data['Close'].ewm(span = days , min_periods = days -1).mean())
    name = 'EWMA_' + str(days) + '_days'
    data[name] = ema
    return data

days_list = [10,50,100,200]
for days in days_list:
    SMA_ = SMA(ADVC,days)
    EWMA_ = EWMA(ADVC,days)
```

```python
SMA_
```

```python
EWMA_
```

```python

#Bollinger Bands (UpperBB & LowerBB)
def bbands(data, window=days):
    MA = data.Close.rolling(window=days).mean()
    SD = data.Close.rolling(window=days).std()
    data['UpperBB'] = MA + (2 * SD) 
    data['LowerBB'] = MA - (2 * SD)
    return data
days = 12
bbands_ = bbands(ADVC, days)
```

```python
bbands_
```

```python
# Force Index (ForceIndex)
def ForceIndex(data, days): 
    FI = pd.Series(data['Close'].diff(days) * data['Volume'], name = 'ForceIndex') 
    data = data.join(FI) 
    return data
days = 1
ForceIndex_ = ForceIndex(ADVC,days)
```

```python
ForceIndex_
```

```python
# Commodity Channel Index (CCI)
def CCI(data, days): 
    TP = (data['High'] + data['Low'] + data['Close']) / 3 
    CCI = pd.Series((TP - TP.rolling(window=days).mean()) / (0.015 * TP.rolling(window=days).std()), name = 'CCI')
    data = data.join(CCI)
    return data
days = 20
CCI_ = CCI(ADVC, days)
```

```python
CCI_.head()
```

```python
# Ease Of Movement (EVM)
def EVM(data, days): 
    dm = ((data['High'] + data['Low'])/2) - ((data['High'].shift(1) + data['Low'].shift(1))/2)
    br = (data['Volume'] / 100000000) / ((data['High'] - data['Low']))
    EVM = dm / br 
    EVM_MA = pd.Series(EVM.rolling(window=days).mean(), name = 'EVM') 
    data = data.join(EVM_MA) 
    return data 
days = 14
EVM_ = EVM(ADVC, days)
```

```python
EVM_.head()
```

```python
# Rate of Change (ROC)
def ROC(data,days):
    N = data['Close'].diff(days)
    D = data['Close'].shift(days)
    roc = pd.Series(N/D,name='ROW')
    data = data.join(roc)
    return data 
days = 5
ROC_ = ROC(ADVC,days)
```
