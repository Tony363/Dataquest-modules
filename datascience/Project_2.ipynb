```python
visited = []
colors = ['red'] * G.number_of_nodes()

def dfs(node):
    if not node in visited:
        print("visited",visited)
        colors[node] = 'blue'
        nx.draw(G, pos, node_color=colors, with_labels=True)
        plt.show()
        visited.append(node)
        neighbors = nx.neighbors(G,node)
        #print(list(neighbors))
        for neighbor in neighbors:
            dfs(neighbor)
```

```python
def BFS(G, node):

    visited = [node]
    depth_limit = len(G)
    queue = deque([(node, depth_limit, nx.neighbors(G, node))])
    while queue:
        parent, depth_now, children = queue[0]
        try:
            child = next(children)
            if child not in visited:
                #print('visited :', visited)
                yield child
                visited.append(child)
                if depth_now > 1:
                    queue.append((child, depth_now - 1, nx.neighbors(G, child)))
```

```python
sorted_edges = sorted(G.edges(data=True), key=lambda x: x[2]['Weight'])
```

```python
sorted_edges
```

```python
Kruskal_G = nx.Graph()
for node1, node2, weight in sorted_edges:
    Kruskal_G.add_edge(node1, node2, weight=weight["Weight"])
    cycles = nx.algorithms.cycles.cycle_basis(Kruskal_G)
    #if not nx.algorithms.tree.recognition.is_tree(Kruskal_G):
    print(cycles)
    if len(cycles) > 0:
        Kruskal_G.remove_edge(node1, node2)
    if Kruskal_G.number_of_edges() == G.number_of_nodes()-1:
        break
```
